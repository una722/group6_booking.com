# Data Processing/Cleaning

```python
#New File
List_clean = List
```

Checking whether there is missing data in this dataframe

```python
missing_data = List_clean.isnull()
missing_data.head(150)
```

```python
#transorm to Capitalize each letter
List_clean['Hotel Name']=List_clean['Hotel Name'].str.title()
List_clean["Distance"]=List_clean["Distance"].str.lower()
List_clean["Room Description"]=List_clean["Room Description"].str.title()
#replace
List_clean['Count Review'] = List_clean['Count Review'].str.replace('Review','')
List_clean['Distance'] = List_clean['Distance'].str.replace(r'miles from center', '', regex=True)

#convert to numerical value
List_clean['Rating'] = pd.to_numeric(List_clean['Rating'])
List_clean['Price'] = List_clean['Price'].str.replace(r'$', '', regex=True)
List_clean.columns = List_clean.columns.str.replace('Price', 'Price/night ($)')


List_clean
```

```python
List_clean['Distance'] = List_clean['Distance'].str.replace(r'feet from center', '', regex=True)
List_clean
```

```python
#average
avg_1 = List_clean["Price/night ($)"].astype('float').mean(axis = 0)

roundedAvg = "{0:.2f}".format(avg_1)
List_clean['Price/night ($)'].replace(np.nan, roundedAvg, inplace=True)
List_clean
```
